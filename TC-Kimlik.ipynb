```python
def isIDAvailable(no):
    """
    Kimlik numaranızın geçerli olup olmadığını kontrol eder
    """
    no = str(no).replace(" ","")
    
    def digitCheck(no):        
        x = 7 * giveMeIntSum(no, [0,2,4,6,8]) - giveMeIntSum(no, [1,3,5,7])
        return x % 10 == int(no[9])
    
    def giveMeIntSum(stringNumber, numList):
        total = 0
        for index in numList:
            total += int(stringNumber[index])
        return total
    
    
    
    isValid = (no.isnumeric() and len(no) == 11) and no[0] != "0"
     
    if not isValid:
        return False
    
    isDigitCheckOK = digitCheck(no)
    isTotalCorrect = giveMeIntSum(no, [n for n in range(10)]) % 10 == int(no[10])    
   
    
    return isDigitCheckOK and isTotalCorrect
    
    
```

```python
class Teacher:
    """
    name, ID, branch
    """
    
    def __init__(self, name, ID, branch):
        self.name = name
        self.ID = ID
        self.branch = branch        
        print("Teacher created:", self.name)
    
    
    def __str__(self):
        return self.name
    
```

```python
class Principal(Teacher):
    """
    name, ID, branch, isReisci
    """
    
    def __init__(self, name, ID, branch, isReisci):
        super().__init__(name,ID,branch)
        self.isReisci = isReisci
        print("Principal created:", self.name)
        
    def __str__(self):
        return self.name + " göreve başladı"
```

```python
class School:
    
    """
    öğrenciler, teacherler, principal, name
    """
    
    students = []
    teachers = []
    principal = None
    
    def __init__(self, name):
        self.name = name
        print("Okullar açıldı\n" + self.name)
        
    def showStudents(self):
        print("Students: ")
        for s in self.students:
            print(s.name)
            
    def showTeachers(self):
        print("Teachers: ")
        for t in self.teachers:
            print(t.name)
            
    def showPrincipal(self):
        print("Principal:", self.principal)
        
    def addStudent(self, student):
        self.students.append(student)
        with open("students.txt", "a") as stds:
            stds.write(str(student) + "\n")
            
    def addTeacher(self, teacher):
        self.teachers.append(teacher)
        with open("teachers.txt", "a") as tchrs:
            tchrs.write(str(teacher) + "\n")
        
    def __str__(self):
        return self.name
```

```python
def getStudent():
    name = input("What's the name of the student: ")
    idNumber = input("Enter the id number: ")
    className = input("Wich class: ")
    
    if not isIDAvailable(idNumber):
        print("ID number is not available")
        return None
    else:
        student = Student(name, idNumber, className)
        return student
```

```python
def getTeacher():
    name = input("What's the name of the teacher: ")
    idNumber = input("Enter the id number: ")
    branch = input("Which brach: ")
    
    if not isIDAvailable(idNumber):
        print("ID number is not available")
        return None
    else:
        teacher = Teacher(name, idNumber, branch)
        return teacher
```

```python
def getPrincipal():
    name = input("What's the name of the teacher: ")
    idNumber = input("Enter the id number: ")
    branch = input("Which brach: ")
    isReisci = 'y' == input("Is reisci? press 'y' for yes 'n' for no" ).lower()
    
    if not isIDAvailable(idNumber):
        print("ID number is not available")
        return None
    else:
        principal = Principal(name, idNumber, branch, isReisci)
        return principal
```

```python
wellcome_message = """
------------FATU EDUCATION--------

Wellcome to our School

Please select an operation:

1- See Students
2- See Teachers
3- See Principal
4- Add Student
5- Add Teacher
6- Change Principal
7- Quit

"""

print(wellcome_message)
operation = 0
while operation!= 7:
    operation = int(input("Please enter: "))
    if operation == 1:
        school1.showStudents()
    
    elif operation == 2:
        school1.showTeachers()
        
    elif operation == 3:
        school1.showPrincipal()
    
    elif operation == 4:
        std = getStudent()
        if std == None:
            print("Student info is not correct")
        else:
            school1.addStudent(std)
            print("New Student is added..", std.name)
    
    elif operation == 5:
        tch = getTeacher()
        if tch == None:
            print("Teacher info is not correct")
        else:
            school1.teachers.append(tch)
            print("New Teacher is added..", tch.name)
            
    elif operation == 6:
        prn = getPrincipal()
        if prn == None:
            print("Principal info is not correct")
        else:
            school1.principal = prn
            print("New Principal is:", prn)
    
        
print("Goodbye")
```
